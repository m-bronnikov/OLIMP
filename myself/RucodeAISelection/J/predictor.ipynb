```python
import numpy as np
import pandas as pd
```

**Take a look at train dataframe**

Import dataframe.

```python
train = pd.read_csv("train.csv")
train.tail()
```

Looks like *target* equal to *idx*. Let's introduce corresponding feature.

```python
train['idx'] = train.index
train.head()
```

Let's check this hypotese.

```python
"Columns are equal:", np.array_equal(train['idx'], train['target'])
```

**Take a look at eval dataframe**

Import data frame and check *target* is equal to *idx*.

```python
eval = pd.read_csv("eval.csv")
eval['idx'] = eval.index
"Columns are equal:", np.array_equal(eval['idx'], eval['target'])
```

**Evaluate**

```python
test = pd.read_csv("test.csv")
result = pd.DataFrame({'target': test.index})
test.head()
```

```python
result.to_csv("result.csv", header=True, index=False)
result.head()
```
